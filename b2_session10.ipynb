```python
tokens = ["the", "animal", "was", "tired"]
attention_scores_for_it = {"the": 0.05, "animal": 0.70, "was": 0.10, "tired": 0.15}
attention_scores_for_it
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.bar(attention_scores_for_it.keys(), attention_scores_for_it.values())
plt.title("Simplified Attention Scores for 'it'")
plt.xlabel("Previous Tokens")
plt.ylabel("Attention Scores")
plt.ylim(0, 1)
plt.show()
```

```python
highest_attention_token = max(attention_scores_for_it, key=attention_scores_for_it.get)
print("Token with highest attention:", highest_attention_token)
```
